## Loading DataSets

```python
it_salary_2018 = pd.read_csv('archive/IT Salary Survey EU 2018.csv')
it_salary_2019 = pd.read_csv('archive/IT Salary Survey EU 2019.csv')
it_salary_2020 = pd.read_csv('archive/IT Salary Survey EU  2020.csv')
```

Joining the data from the three years:<br/>
- 2018
- 2019
- 2020

```python
it_salary = pd.concat([it_salary_2018, it_salary_2019, it_salary_2020], ignore_index=True)
```

Drop the columns

```python
it_salary.columns
```

```python
it_salary = it_salary.drop('Annual brutto salary (without bonus and stocks) one year ago. Only answer if staying in the same country', axis=1)
it_salary = it_salary.drop('Have you been forced to have a shorter working week (Kurzarbeit)? If yes, how many hours per week', axis=1)
it_salary = it_salary.drop('Have you received additional monetary support from your employer due to Work From Home? If yes, how much in 2020 in EUR', axis=1)
it_salary = it_salary.drop('Timestamp', axis=1)
it_salary = it_salary.drop('Yearly bonus + stocks in EUR', axis=1)
it_salary = it_salary.drop('Annual bonus+stocks one year ago. Only answer if staying in same country', axis=1)
it_salary.shape
```

<strong>Rename Columns for Readability</strong>


```python
it_salary.rename(columns = {
    'Total years of experience':'experience',
    'Years of experience in Germany':'germany_experience',
    'Seniority level':'work_level',
    'Your main technology / programming language':'main_tech',
    'Other technologies/programming languages you use often':'other_tech',
    'Number of vacation days':'vacation_days',
    'Yearly brutto salary (without bonus and stocks) in EUR':'yearly_salary',
    'Employment status':'emp_state',
    'Сontract duration':'contract_duration',
    'Main language at work':'main_language',
    'Company size':'company_size',
    'Company type':'company_type',
    'Have you lost your job due to the coronavirus outbreak?':'covid_lost',
    'Age':'age',
    'Gender':'gender',
    'City':'city',
    'Position ':'position'
    
}, inplace = True)
```

### Filling Null Values

```python
it_salary['age'] = it_salary['age'].fillna(it_salary.groupby('experience')['age'].transform('mean'))


#drop rows with null values and null experience
it_salary.dropna(subset=['age'], how='all', inplace=True)

#Changing the age datatype to int
```
